### Paddding

Define method named pad_with_zeros that performs specified number of zero padding on the input data.
- parameters:
    - X: input data
    - pad: number of zero padding
- returns:
    - data_padded: the nd-array after padding  


(O padding adiciona zeros ao redor das bordas de uma imagem antes de aplicar operaes de convoluo. Isto  feito para preservar as dimenses da imagem e evitar a reduo excessiva do tamanho durante o processo de convoluo)

```python
def pad_with_zeros(data: np.ndarray, pad: int) -> np.ndarray:
    """
    Pad image with zeros. The padding is applied to the height and width of an image,
    
    Parameters
    ----------
    data: numpy array
        input data
    pad: int
        number of zero padding

    Returns
    -------
    data_padded: numpy array
        the nd-array after padding
    """
    # podemos aplicar padding com zeros usando np.pad
    data_padded = np.pad(data, ((0, 0), (pad, pad), (pad, pad), (0, 0)), mode="constant", constant_values=0)
    return data_padded

# a funo numpy.pad recebe o array original, o nmero de elementos a serem adicionados ao incio e ao final de cada dimenso (parmetro pad_width), 
# o modo de padding (neste caso, 'constant' para preencher com zeros) e os valores constantes a serem usados para o preenchimento (neste caso, 0). 
```

```python
padded_image = pad_with_zeros(data, pad=10)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.title('Imagem Original')
plt.imshow(data[0])  # Mostrar apenas a primeira (e nica) imagem na dimenso 0
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagem com Padding de Zeros')
plt.imshow(padded_image[0])  # Mostrar apenas a primeira (e nica) imagem na dimenso 0
plt.axis('off')

plt.show()
```

### Convolution

#### Single step convolution

Define the function named single_step_convolution() to convolve a slice of input data using the specified filter.

- parameters:

   - data_slice: the slice of input data
   - W: the filter used for convolution  
   - b: the bias term
     
- returns:

   - Z: the output of the convolution operation

```python
def single_step_convolution(data_slice: np.ndarray, W: np.ndarray, b: np.ndarray) -> np.ndarray:
    """
    Convolve a slice of input data using the specified filter and bias.
    
    Parameters
    ----------
    data_slice: numpy array
        the slice of input data
    W: numpy array
        the filter used for convolution
    b: numpy array
        the bias term
        
    Returns
    -------
    Z: numpy array
        the output of the convolution operation
    """
    Z = np.sum(data_slice * W) + float(b)  #sum of the input data with the filter, and multiplied by bias term
    return Z
```

#### Strided convolution

Strided convolution is a variation of the convolution operation in CNNs where the filter does not move one pixel at a time but skips a certain number of pixels, known as the stride, during each step of the convolution.

Define method convolution_forward to perform strided convolution on the input data. Use single_step_convolution() to perform the convolution at each stride.

- parameters:

   - data: the input data
   - W: the filter used for convolution
   - b: the bias term
   - stride: the stride used for convolution
   - pad: the number of zero padding used for convolution
  
- returns:
    - Z: the output of the convolution operation

```python
def convolution_forward(data: np.ndarray, W: np.ndarray, b: np.ndarray, stride: int, pad: int) -> np.ndarray:
    """
    Perform strided convolution on the input data.
    
    Parameters
    ----------
    data: numpy array
        the input data
    W: numpy array
        the filter used for convolution
    b: numpy array
        the bias term
    stride: int
        the stride used for convolution
    pad: int
        the number of zero padding used for convolution

    Returns
    -------
    Z: numpy array
        the output of the convolution operation
    """
    # dimensions are m (number of examples), h_prev (height), w_prev (width), c_prev (number of channels)
    m, h_prev, w_prev, c_prev = data.shape  # samples, height, width, chanel (RGB usually)

    # Dimensions of the filter
    f1, f2, c_prev, n_c = W.shape  # height, width of the filter, chanels, number of filters (output chanels)

    # Calculate dimensions of the output AFTER CONVOLUTION (input_size - filter size + 2 * padding / stride + 1) (+2 is related to add in heigth or width the image, the +1 guarantees that the output covers completely the entry)
    n_h = int((h_prev - f1 + 2 * pad) / stride) + 1
    n_w = int((w_prev - f2 + 2 * pad) / stride) + 1

    # Initialize the output Z with zeros
    Z = np.zeros((m, n_h, n_w, n_c))  # samples, output_heigth, output_width, chanels

    # Pad the input data
    A_prev_pad = np.pad(data, ((0, 0), (pad, pad), (pad, pad), (0, 0)), mode='constant', constant_values=0)
    #adds padding_zero, with data and dimensions of the padding, mode constant to fill with zeros

    for i in range(m):  # four loops for all dimensions (sample, height, width, chanel)
        for h in range(n_h):
            for w in range(n_w):
                for c in range(n_c):
                    # Find the slice of the input data to be convolved
                    w_start = w * stride  # 
                    w_end = w_start + f2
                    h_start = h * stride
                    h_end = h_start + f1

                    Z[i, h, w, c] = single_step_convolution(A_prev_pad[i, h_start:h_end, w_start:w_end, :], W[:, :, :, c], b[:, :, :, c])  #saving the result of the matrix

    return Z
```

```python
np.random.seed(1)
# dimensions are 10 batch size, 4 height, 4 width, 3 channels
input_ = np.random.randn(10, 4, 4, 3)
# dimensions are 2 height, 2 width, 3 channels, 8 filters
W = np.random.randn(2, 2, 3, 8)
# dimensions are 1, 1, 1, 8, i.e. one bias per filter
b = np.random.randn(1, 1, 1, 8)
pad = 1
stride = 1

output_ = convolution_forward(input_, W, b, stride, pad)
print(np.mean(output_))
```

### Pooling

#### Max pooling

Define method max_pooling to perform max pooling on the input data.

- parameters:

    - data: input data on which convolution is performed  
    - f: filter size 
    - stride: the number of strides
    
- returns:
    - Z: the output of the max pooling operation

```python
Image(url='data/maxpool.gif')
```

```python
def max_pooling(input: np.ndarray, f: int, stride: int) -> np.ndarray:
    """
    Perform max pooling on the input data.
    
    Parameters
    ----------
    input: numpy array
        input data on which convolution is performed
    f: int
        filter size
    stride: int
        the number of strides

    Returns
    -------
    output: numpy array
        the output of the max pooling operation
    """
    m, h_prev, w_prev, c_prev = input.shape

    # calculate the dimension of the output (h_out is the height of the output, w_out is the width of the output)
    h_out = int((h_prev - f) / stride) + 1  #here there is no padding so theres no need to add * 2
    w_out = int((w_prev - f) / stride) + 1

    # initialize the output Z with zeros (4 dimensions -> batch size, height, width, number of channels)
    output = np.zeros((m, h_out, w_out, c_prev))  # output array initialized with zeros. The dimensions match the number of examples in the batch
    
    # nested loops to iterate through each example in the batch (i), each channel (c), and each position in the output tensor (h and w)
    for i in range(m):
        for c in range(c_prev):
            for h in range(h_out):
                for w in range(w_out):
                    # find the slice of the input data to be pooled (take into account the stride and filter size (f))
                    # calculating the starting and ending indices to define the region in the input data to be pooled
                    # these indices are based on the current position in the output tensor, the filter size (f), and the stride (stride)
                    w_start = w * stride
                    w_end = w_start + f
                    h_start = h * stride
                    h_end = h_start + f
                    output[i, h, w, c] = np.max(input[i,h_start:h_end, w_start:w_end, c])  #find the maximum value in the specified region of the input data and assigns it to the corresponding position in the output tensor
    assert output.shape == (m, h_out, w_out, c_prev)  #assertion to ensure that the shape of the output tensor matches the expected shape
    return output
```
