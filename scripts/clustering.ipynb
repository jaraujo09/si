```python
# imports
import matplotlib.pylab as plt
import numpy as np
from si.io.csv_file import read_csv
from si.clustering.kmeans import KMeans
```

```python
# datasets
filename= r'C:\Users\Fofinha\Desktop\UNI\MESTRADO\2o ANO\Sistemas Inteligentes\si\datasets\iris\iris.csv'
iris_dataset = read_csv(filename=filename, features=True, label=True)
```

```python
# kmeans
k = 3
kmeans = KMeans(k)
distances = kmeans.fit_transform(iris_dataset)
predictions = kmeans.predict(iris_dataset)
predictions
```

```python
# plotting
for i in range(k):
    k_sample = iris_dataset.X[np.where(predictions == i)]
    plt.plot(k_sample[:, 0], k_sample[:, 1], 'o')
    lines = plt.plot(kmeans.centroids[i, 0], kmeans.centroids[i, 1], 'kx')
    plt.setp(lines, ms=15.0)
    plt.setp(lines, mew=2.0)

plt.show()
```
